```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Airline Dataset1.xlsx', sheetname='Airline Dataset1')
df.head()
```

```python
df1 = np.asanyarray(df[['time','adv_bookdays','demand','Class']])
df2 = np.asanyarray(df['Fare'])
```

```python
df.plot(kind ='scatter', x = 'demand', y = 'Fare')
plt.show
```

```python
df.plot(kind ='scatter', x = 'time', y = 'Fare')
plt.show
```

```python
df.plot(kind ='scatter', x = 'adv_bookdays', y = 'Fare')
plt.show
```

```python
df.corr()
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
reg.fit(df1,df2)
```

```python
reg.coef_
```

```python
test = pd.read_excel('Airline Dataset1.xlsx', sheetname= 'TestDataSet')
test.head()
```

```python
test1 = np.asanyarray(test[['time','adv_bookdays','demand','Class']])
```

```python
test['Fare'] = reg.predict(test1)
```
